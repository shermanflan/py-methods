```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
df1 = pd.read_csv('./data/concat_1.csv')
df2 = pd.read_csv('./data/concat_2.csv')
df3 = pd.read_csv('./data/concat_3.csv')

df1
```

# Concatenation

```python
# With re-indexing
row_concat = pd.concat([df1, df2, df3], ignore_index=True)

row_concat
```

```python
new_row_df = pd.DataFrame(
                [['a12', 'b12', 'c12', 'd12']] # double bracket for DF
                , columns=['A', 'B', 'C', 'D']
            )
row_concat = pd.concat(
                objs=[row_concat, new_row_df]
                , ignore_index=True
            )

row_concat
```

```python
col_concat = pd.concat(
                [df1, df2, df3]
                , ignore_index=True
                , axis='columns'
            )
col_concat.columns = ['A', 'B', 'C', 'D'
                      , 'E', 'F', 'G', 'H'
                      , 'I', 'J', 'K', 'L']
col_concat
```
